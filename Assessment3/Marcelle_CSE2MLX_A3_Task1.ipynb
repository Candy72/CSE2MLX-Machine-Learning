```python
import matplotlib.pyplot as plt
```

```python
# Original Points categorised by their new cluster assignments
cluster_1_points = [(5,0), (3,3), (2,4)]
cluster_2_points = [(7,9), (3,5), (7,6)]

# Centroids
centroids = [(4.0,1.5), (4.75, 6.0)]
```

```python
# Plotting Cluster 1 points
for point in cluster_1_points:
    plt.scatter(*point, color='blue')

# Plotting Cluster 2 points
for point in cluster_2_points:
    plt.scatter(*point, color='red')

# Plotting Centroids
for centroid in centroids:
    plt.scatter(*centroid, color='green', marker='x', s=100) 

plt.title('Cluster visualisation with new centroids')
plt.xlabel('X coordinate')
plt.ylabel('Y coordinate')
plt.show()
```

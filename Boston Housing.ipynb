## Split data

- Note that data hasn't been modified until now. 

```python
# Select feature
selected_columns = ['LSTAT', 'RM', 'PTRATIO', 'INDUS', 'TAX']
final_df = boston_df.loc[:, selected_columns]

# Display the new DataFrame
print(final_df.head())
```

```python
# Note X (Uppercase) and y (lowercase)
X = final_df
y = boston_df['MEDV']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3,
random_state=5)
```

```python
print(X_train.shape)
print(y_train.shape)
```

```python
print(X_test.shape)
print(y_test.shape)
```

### STOP: have you seen any scaling here yet?

## Train Model

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

## Predict

```python
price_pred = model.predict(X_test)
```

```python
print(price_pred)
```

## Evaluate Model

```python
# print('R-Squared: %.4f' % model.score(X_test, y_test)) # if you know formating
print('R-Squared:',model.score(X_test, y_test))
```

```python
from sklearn.metrics import mean_squared_error
mse = mean_squared_error(y_test, price_pred)
print(mse)
plt.scatter(y_test, price_pred)
plt.xlabel("Actual Prices")
plt.ylabel("Predicted prices")
plt.title("Actual prices vs Predicted prices")
```

## Extra for Model Evaluation (not from the book)

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
```

```python
mse = mean_squared_error(y_test, price_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, price_pred)
r_squared = r2_score(y_test, price_pred)

print("Mean Squared Error (MSE):", mse)
print("Root Mean Squared Error (RMSE):", rmse)
print("Mean Absolute Error (MAE):", mae)
print("R-squared (R2):", r_squared)
```

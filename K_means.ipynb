```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os
os.environ['OMP_NUM_THREADS'] = '1'
```

```python
dataset = pd.read_csv ('datasets/online_customers.csv')
X = dataset.values
```

```python
from sklearn.cluster import KMeans
wcss = []
for i in range (1,11):
    kmeans = KMeans(n_clusters=i, n_init=10, random_state=0).fit(X)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('The Elbow Methods')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.show
```

```python
kmeans = KMeans(n_clusters=5, n_init=10, random_state=0).fit(X)
kmeans.fit(X)
y_kmeans = kmeans.predict(X)
print (y_kmeans)
```

```python
plt.scatter(X[y_kmeans == 0,0], X[y_kmeans == 0,1], s = 100, c = 'red', label = 'Cluster 1')
# plt.scatter(X[y_kmeans == 1,0], X[y_kmeans == 1,1], s = 100, c = 'blue', label = 'Cluster 2')
# plt.scatter(X[y_kmeans == 2,0], X[y_kmeans == 2,1], s = 100, c = green, label = 'Cluster 3')
# plt.scatter(X[y_kmeans == 3,0], X[y_kmeans == 3,1], s = 100, c = 'black', label = 'Cluster 4')
# plt.scatter(X[y_kmeans == 4,0], X[y_kmeans == 4,1], s = 100, c = 'grey', label = 'Cluster 5')
# plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 300, c = 'yellow', label = 'Centroids')
plt.title('Customer Segments')
plt.xlabel('Number of Visits')
plt.ylabel('Number of purchases items')
plt.legend()
plt.show
```
